```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text: str):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    tokens = nltk.word_tokenize(text)
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens

print(preprocess_text(song1))
```

```python
songs_clean = [" ".join(preprocess_text(text)) for text in songs]
for name, txt in zip(song_names, songs_clean):
    print(f"{name}: {txt}")
```

After pre-processing, there were significantly fewer words left in the texts, with pronouns and function words disappearing.
This facilitates the analysis of meaningful lexical units and reduces the size of the dictionary.
Lemmatisation made it possible to reduce different forms of words to a single base, thanks to which the frequency of keywords is considered more accurate.

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(songs_clean)
feature_names = tfidf_vectorizer.get_feature_names_out()

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),
                        index=song_names,
                        columns=feature_names)
tfidf_df
```

```python
def top_tfidf_words(row, feature_names, top_n=10):
    row_arr = row.toarray().flatten()
    top_idx = np.argsort(row_arr)[::-1][:top_n]
    return [(feature_names[i], row_arr[i]) for i in top_idx]

for idx, name in enumerate(song_names):
    print(f"\n=== {name} ===")
    for word, score in top_tfidf_words(tfidf_matrix[idx], feature_names, top_n=5):
        print(f"{word:15s} {score:.3f}")
```

```python
from gensim.models import Word2Vec
corpus_tokens = [preprocess_text(text) for text in songs]

w2v_model = Word2Vec(
    sentences=corpus_tokens,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4,
    sg=1  # skip-gram
)

def song_vector(tokens, model):
    vecs = []
    for t in tokens:
        if t in model.wv:
            vecs.append(model.wv[t])
    if len(vecs) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vecs, axis=0)

song_vectors = np.vstack([song_vector(tokens, w2v_model) for tokens in corpus_tokens])
song_vectors.shape
```

CountVectorizer is simple to implement and well suited for basic statistics, but does not take into account the importance of words in different documents.
TF-IDF provides more informative weights and is often better suited for classical ML models.
Word2Vec provides dense vectors that reflect the meaning of words well, but requires training and is more difficult to interpret.

## Part 2: Sentiment Classification with BERT

```python
import pandas as pd
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
label_map = {'positive': 1, 'pos': 1, 'negative': 0, 'neg': 0, 1: 1, 0: 0}
df['label'] = df['sentiment'].map(label_map)
df = df.dropna(subset=['label'])
df['label'] = df['label'].astype(int)
df[['review', 'label']].head()
```

```python
from sklearn.model_selection import train_test_split
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['review'].tolist(),
    df['label'].tolist(),
    test_size=0.2,
    random_state=42,
    stratify=df['label'])
```

```python
import torch
from transformers import BertTokenizerFast, BertForSequenceClassification
model_name = "bert-base-uncased"
tokenizer = BertTokenizerFast.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
device
```

```python
max_len = 64
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=max_len)

test_encodings = tokenizer(
    test_texts,
    truncation=True,
    padding=True,
    max_length=max_len)
```

```python
class IMDBDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

train_dataset = IMDBDataset(train_encodings, train_labels)
test_dataset = IMDBDataset(test_encodings, test_labels)
```

```python
from torch.utils.data import DataLoader
from torch.optim import AdamW
from tqdm.auto import tqdm
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
optim = AdamW(model.parameters(), lr=2e-5)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def train_epoch(model, data_loader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in tqdm(data_loader, desc="Training"):
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(data_loader)


def eval_model(model, data_loader, device):
    model.eval()
    preds = []
    true_labels = []

    with torch.no_grad():
        for batch in tqdm(data_loader, desc="Evaluating"):
            labels = batch['labels'].to(device)
            inputs = {k: v.to(device) for k, v in batch.items() if k != 'labels'}

            outputs = model(**inputs)
            logits = outputs.logits
            batch_preds = torch.argmax(logits, dim=1)

            preds.extend(batch_preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    acc = accuracy_score(true_labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(
        true_labels, preds, average='binary', pos_label=1)
    return acc, precision, recall, f1
```

```python
num_epochs = 4
for epoch in range(num_epochs):
    print(f"\n===== Epoch {epoch+1}/{num_epochs} =====")
    train_loss = train_epoch(model, train_loader, optim, device)
    acc, prec, rec, f1 = eval_model(model, test_loader, device)
    print(f"Train loss: {train_loss:.4f}")
    print(f"Test accuracy: {acc:.4f}")
    print(f"Test precision: {prec:.4f}")
    print(f"Test recall: {rec:.4f}")
    print(f"Test F1: {f1:.4f}")
```

The model showed the best quality after the second epoch (accuracy ≈ 0.8545, F1 ≈ 0.8549).
A further increase in the number of epochs leads to a decrease in quality on the test sample
with a continuing decrease in the training loss function, which indicates the onset of overfitting.
Precision and recall for the positive class are approximately equal, which indicates a good balance
between missed and false positive examples.

```python
from sklearn.metrics import classification_report

def get_predictions_and_labels(model, data_loader, device):
    model.eval()
    preds = []
    true_labels = []

    with torch.no_grad():
        for batch in data_loader:
            labels = batch['labels'].to(device)
            inputs = {k: v.to(device) for k, v in batch.items() if k != 'labels'}

            outputs = model(**inputs)
            logits = outputs.logits
            batch_preds = torch.argmax(logits, dim=1)

            preds.extend(batch_preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())
    return preds, true_labels

preds, labels_true = get_predictions_and_labels(model, test_loader, device)
print(classification_report(labels_true, preds, target_names=['negative', 'positive']))
```

```python
def predict_review_sentiment(text, model, tokenizer, device, max_len=64):
    model.eval()
    encoding = tokenizer(
        text,
        truncation=True,
        padding=True,
        max_length=max_len,
        return_tensors='pt'
    )
    encoding = {k: v.to(device) for k, v in encoding.items()}

    with torch.no_grad():
        outputs = model(**encoding)
        logits = outputs.logits
        probs = torch.softmax(logits, dim=1).cpu().numpy()[0]
        pred_label = np.argmax(probs)

    return pred_label, probs

sample_text = "This movie was absolutely wonderful! The acting and story were amazing."
label, probs = predict_review_sentiment(sample_text, model, tokenizer, device)
print("Predicted label:", "positive" if label == 1 else "negative")
print("Probabilities [neg, pos]:", probs)
```
